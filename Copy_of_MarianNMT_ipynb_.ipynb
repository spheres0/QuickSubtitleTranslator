```python
from transformers import MarianMTModel, MarianTokenizer

# Модель для перевода с английского на русский
model_name = 'Helsinki-NLP/opus-mt-en-ru'

# Загружаем модель и токенизатор
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
```

```python
def translate_text(text, model, tokenizer):
    # Токенизация текста
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    # Перевод текста
    translated = model.generate(**inputs)

    # Декодируем результат
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

    return translated_text
```

```python
import pysrt
from google.colab import files

# Загружаем файл субтитров
uploaded = files.upload()  # Загружаем файл .srt через интерфейс Colab
file_name = list(uploaded.keys())[0]  # Получаем имя загруженного файла

# Открываем субтитры
subs = pysrt.open(file_name)

# Переводим все субтитры
for sub in subs:
    sub.text = translate_text(sub.text, model, tokenizer)  # Переводим каждый субтитр

# Сохраняем переведённые субтитры
output_file = file_name.replace('.srt', '_translated.srt')
subs.save(output_file)

# Скачиваем файл с переведёнными субтитрами
files.download(output_file)
```
